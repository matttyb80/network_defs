```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node('0x007',provider= 'Data Provider 0x01', record= 'record 0x007')
G.add_node('0x008',provider= 'Data Provider 0x01', record= 'record 0x008')
G.add_node('0x009',provider= 'Data Provider 0x02', record= 'record 0x009')
```

```python
G.add_edge('0x007', '0x008', label="Query 1")
G.add_edge('0x007', '0x009', label="Query 2")
```

```python
plt.figure(figsize=(6, 4))
nx.draw(G, pos=nx.circular_layout(G), node_size=1300, node_color=['orange','orange','cyan'], with_labels=True)
nx.draw_networkx_edge_labels(
    G, pos=nx.circular_layout(G), edge_labels={('0x007', '0x008'):'Query 1',('0x007', '0x009'):'Query 2'},
    font_color='black'
)
plt.savefig("sample_network.png", bbox_inches='tight', transparent=True)
plt.show()
```
